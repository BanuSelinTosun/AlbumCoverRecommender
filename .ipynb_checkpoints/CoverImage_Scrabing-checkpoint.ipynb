```python
img_src = []
for elem in album_image_list:
    val = elem['src']
    cur_src = (val, elem['data-src'].split('/')[7].split('photo')[0])
    if not val.endswith('s_1x2.gif'):
        img_src.append(cur_src)
img_src
```

```python
add_on = 'album_images'
save_dir = './album_covers/'
os.makedirs(save_dir)
images_path = '{}/{}'.format(save_dir, add_on)
os.makedirs(images_path)
for path, name in img_src:
    i = Image.open(requests.get(path, stream=True).raw)
    cur_path='{}/{}{}.{}'.format(images_path, name, path.split('/')[-1], 'jpeg')
    i.save(cur_path)
    print cur_path
```

```python
npage = 'https://www.discogs.com/lists/Rolling-Stones-100-Greatest-Album-Covers/25102?page=1&limit=250'
nr = requests.get(npage)
nr.status_code
```

```python
client = MongoClient('mongodb://localhost:27017/')
ndb = client.metroid
pages = ndb.pages

pages.insert_one({'html': nr.content})
```

```python
nsoup = BeautifulSoup(nr.content)
print(nsoup.prettify())
```

```python
album_list = nsoup.select('.listitem_image')
album_list
```

```python
album_list[0]#['img']
```

```python
for elem in album_list:
    tag = elem.select('img')[0]
    val = tag['src']
    #cur_src = (val, elem['href'].split('/')[1])
    print val
```

```python
album_img_src = []
for elem in album_list:
    tag = elem.select('img')[0]
    val = tag['src']
    ref = str(elem)
    href = ref.split('href')[1]
    name = ' '.join(str(href.split('/')[1]).split('-'))
    cur_src = (val, name)
    album_img_src.append(cur_src)
album_img_src
```

```python
add_on = 'RS100_album_images'
save_dir = './RS100_album_covers/'
os.makedirs(save_dir)
images_path = '{}/{}'.format(save_dir, add_on)
os.makedirs(images_path)
for path, name in album_img_src:
    i = Image.open(requests.get(path, stream=True).raw)
    cur_path='{}/{}{}'.format(images_path, name, path.split('/')[-1])
    i.save(cur_path)
    print cur_path
```

```python
add_on = 'RS100_album_images'
save_dir = './RS100_album_covers/'
images_path = '{}/{}'.format(save_dir, add_on)
add_on_blr = 'blurred_RS100_album_images'
save_dir_blr = './blurred_RS100_album_covers/'
os.makedirs(save_dir_blr)
images_path_blr = '{}/{}'.format(save_dir_blr, add_on_blr)
os.makedirs(images_path_blr)
for path, name in album_img_src:
    cur_path='{}/{}{}'.format(images_path, name, path.split('/')[-1])
    blr_cur_path='{}/blurred_{}{}'.format(images_path_blr, name, path.split('/')[-1])
    trial = Image.open(cur_path)
    blurry = trial.filter(ImageFilter.GaussianBlur)
    blurry.save(blr_cur_path)
```
